```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(x):
    return np.maximum(0.0, x)
```

```python
def leaky_relu(x, alpha):
    return np.maximum(alpha * x, x)
```

```python
x = np.linspace(-10, 10, 1000)
y = relu(x)
```

```python
plt.plot(x, y, label='ReLU')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Rectified Linear Unit')
plt.legend()
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)
y = leaky_relu(x, 0.3)
```

```python
plt.plot(x, y, label='Leaky ReLU')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Leaky Rectified Linear Unit (alpha=0.3)')
plt.legend()
plt.show()
```
